```python
from torchsummary import summary
import sys

# 출력 파일 설정
original_stdout = sys.stdout  # 현재 stdout 백업
with open("model_summary.txt", "w") as f:
    sys.stdout = f  # stdout을 파일로 설정
    summary(model_cpu, input_size=(3, 299, 299), device="cpu")  # Summary 출력
    sys.stdout = original_stdout  # stdout 복구
```

```python
from torchinfo import summary

# 모델 구조 출력 (detail을 높이면 더 많은 정보 출력)
summary(model_cpu, input_size=(1, 3, 299, 299), col_names=["input_size", "output_size", "num_params"], depth=5)
```

# GPU

```python
# GPU가 사용 가능한지 확인
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# 모델을 GPU로 이동
model_gpu = InceptionV2(num_classes=1000).to(device)

# GPU에서 Summary 출력
summary(model_gpu, input_size=(3, 299, 299), device=str(device))
# Summary를 파일로 저장
with open("gpu_model_summary.txt", "w") as f:
    original_stdout = sys.stdout
    sys.stdout = f  # stdout을 파일로 설정
    summary(model_gpu, input_size=(3, 299, 299), device=str(device))
    sys.stdout = original_stdout  # stdout 복구
```
